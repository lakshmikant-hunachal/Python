```python
import numpy as np

A = np.array([[1, 2],
              [3, 4],
              [5, 6]])

b = np.array([1, 3, 5])

AT_A = A.T @ A
AT_b = A.T @ b

x = np.linalg.solve(AT_A, AT_b)

print("Solution x =")
print(x)
```

```python
import numpy as np
from sklearn.decomposition import PCA

X = np.array([[2.5, 2.4, 0.5],
              [0.5, 0.7, 3.2],
              [2.2, 2.9, 1.1]])

pca = PCA(n_components=2)

X_reduced = pca.fit_transform(X)

print("Reduced Dataset:")
print(X_reduced)

print("\nPercentage of Variance Explained:")
print(pca.explained_variance_ratio_ * 100)
```
