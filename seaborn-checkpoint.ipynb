```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
d={"Days":[1,2,3,4,5,6,7],
  "NO of people":[45,52,48,62,56,89,101]}
df=pd.DataFrame(d)
df
```

```python
sns.lineplot(x="Days",y="NO of people",data=df)
plt.show()
```

```python
d={"Total_bill":[16.99,10.34,21.01,23.68,24.59],
   "tip":[1.01,1.66,3.50,3.31,3.61],
   "sex":["Female","Male","Male","Male","Female"],
   "smoker":["No"]*5,
   "Time":["Dinner"]*5,
   "size":[2,3,3,2,4]}
df1=pd.DataFrame(d)
df1
```

```python
sns.lineplot(x="Total_bill",y="size",data=df1,hue="sex")
```

```python
#barplot using seaborn 
df1
```

```python
sns.barplot(x="sex",y="Total_bill",data=df1)
plt.show()
```

```python
#ci varible remove the line on the box
sns.barplot(x="sex",y="Total_bill",data=df1,ci=False)
plt.show()
```

```python
data={"City":["Vijayapur","Belagavi","mysore","Bengaluru","hasan"],
      "number":[40000,50000,25000,30000,78000]
     }
df=pd.DataFrame(data)
df
```

```python
sns.barplot(x="City",y="number",data=df,facecolor=[1,1,1,0],edgecolor="0.00001")
plt.grid()
plt.show()
```

```python
sns.barplot(x="City",y="number",data=df)
plt.show()
```

```python
sns.barplot(y="City",x="number",data=df)
plt.grid()
plt.legend()
plt.show()
```

```python
#Histogram in seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
d={"Total_bill":[16.99,10.34,21.01,23.68,24.59],
   "tip":[1.01,1.66,3.50,3.31,3.61],
   "sex":["Female","Male","Male","Male","Female"],
   "smoker":["No"]*5,
   "Time":["Dinner"]*5,
   "size":[2,3,3,2,4]}
df=pd.DataFrame(d)
df
```

```python
sns.histplot(data= df,kde="sex",linestyle="--)
```

```python
sns.histplot(x="Total_bill",data= df,kde="sex",palette="spring",linestyle="--")
```
