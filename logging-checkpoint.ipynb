```python
import logging as log
#logging is help for n to n function program helps its trace error,and etc
#Basic config of the logging
log.basicConfig(level=log.DEBUG)
log.debug("This is a debug statement")
log.warning("This is a worning statement")
log.critical("This ia a critical statement")
log.info("This is a info statement")
log.error("This is a error statement")
```

```python
import logging as log
log.basicConfig(
    filename="App.log",
    filemode="w",
    level=log.DEBUG,
    format="%(asctime)s-%(name)s-%(levelname)s-%(message)s",
    datefmt="%m-%d-%Y %H:%M:%S"
)

log.debug("This is a debug statement")
log.warning("This is a worning statement")
log.critical("This ia a critical statement")
log.info("This is a info statement")
log.error("This is a error statement")
log.debug("My name is lakshmikant")
log.warning("my father name i sgiramallappa")
log.critical("This ia a critical statement")
log.info("This is a info statement")
log.error("This is a error statement")
```
