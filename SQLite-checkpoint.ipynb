```python
import sqlite3 as sq
connection=sq.connect("Example.db")
connection
```

```python
cursor=connection.cursor()
```

```python
#Create table
cursor.execute(''' create table if not exists student(
USN Integer Primary key,
name Text Not Null,
age Integer,
dept Text
)
''')
##commit a change
connection.commit()
```

```python
cursor.execute(''' 
select * from student
''')
```

```python
##insert data into table from SQLite3
cursor.execute('''
insert into student(USN,name,age,dept)
values(115,"Lakshmikant",20,"CSE")''')

cursor.execute('''
insert into student(USN,name,age,dept)
values(116,"mallu",21,"CIVIL")''')

cursor.execute('''
insert into student(USN,name,age,dept)
values(110,"deepa",25,"ISE")''')

cursor.execute('''
insert into student(USN,name,age,dept)
values(121,"gayatri",22,"MECH")''')

##commit a change
connection.commit()
```

```python
##query the data from the table
cursor.execute('select * from student')
rows=cursor.fetchall()
for row in rows:
    print(row)
```

```python
cursor.execute('''
update student
set age=47
where name="Lakshmikant"
''')
connection.commit()
```

```python
cursor.execute('select * from student')
rows=cursor.fetchall()
for row in rows:
    print(row)
```

```python
import sqlite3 as sq
connection=sq.connect("school_data.db")
cursor=connection.cursor()
connection.commit()
```
