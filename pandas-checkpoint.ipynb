```python
#concatation of two dataframe using merge differance between merge and concat method merge method is remove the same columns
df1=pd.DataFrame({"city":["bagalakote","ilakal","jamakandi"],
                  "temp":[36,37,40],})
df2=pd.DataFrame({"city":["bagalakote","jamakandi","ilakal"],
                  "humidity":[86,92,88]})
df=pd.merge(df1,df2,on="city")
df
```

```python
#add some different cities name to the both dataframe
df1=pd.DataFrame({"city":["bagalakote","ilakal","jamakandi","navanagar"],
                  "temp":[36,37,40,45],})
df2=pd.DataFrame({"city":["bagalakote","jamakandi","ilakal","vijayapur"],
                  "humidity":[86,92,88,89]})
df=pd.merge(df1,df2,on="city",how="inner")
df
#default how value is inner join
```

```python
df1=pd.DataFrame({"city":["bagalakote","ilakal","jamakandi","navanagar"],
                  "temp":[36,37,40,45],})
df2=pd.DataFrame({"city":["bagalakote","jamakandi","ilakal","vijayapur"],
                  "humidity":[86,92,88,89]})
df=pd.merge(df1,df2,on="city",how="outer")
df
#outer join means union function
```

```python
df1=pd.DataFrame({"city":["bagalakote","ilakal","jamakandi","navanagar"],
                  "temp":[36,37,40,45],})
df2=pd.DataFrame({"city":["bagalakote","jamakandi","ilakal","vijayapur"],
                  "humidity":[86,92,88,89]})
df=pd.merge(df1,df2,on="city",how="left")
df
#left join works on the how we will passes to the merge
```

```python
df1=pd.DataFrame({"city":["bagalakote","ilakal","jamakandi","navanagar"],
                  "temp":[36,37,40,45],})
df2=pd.DataFrame({"city":["bagalakote","jamakandi","ilakal","vijayapur"],
                  "humidity":[86,92,88,89]})
df=pd.merge(df1,df2,on="city",how="right")
df
#right join same as left
```
