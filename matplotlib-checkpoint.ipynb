```python
#countplot
import matplotlib.pyplot as pt
import pandas as pd
%matplotlib inline
data={
    "Salary":[25000,26000,28000,30000,35800,56400,41000,45000,48000,45000,35000,24000]
}
df=pd.DataFrame(data)
df["dept"]=["HR","IT","Non IT","HR","Non IT","HR"]*2
count=df["dept"].value_counts()
count
pt.bar(count.index,count,color=["red","black","green"])
#upto work on only one varible this process is called univarite method
```

```python
#when table or dataframe contain tw numeric value column solve using bivarite method
#first method is  scatter method
df
df["Age"]=[24,25,42,35,26,28,24,31,26,32,27,35]
df
```

```python
#scatter method
pt.scatter(df["Salary"],df["Age"],color="red")
pt.grid()
```

```python
#second method is lineplot method
pt.plot(df["Salary"],df["Age"],marker="*")
#graph is some more complex can not read easy due to numeric values are not sorted
pt.grid()
pt.show()
```

```python
import matplotlib.pyplot as pt

sort_age = df.sort_values("Age")

pt.plot(sort_age["Age"], sort_age["Salary"], marker="*")
pt.grid()
pt.show()
```
