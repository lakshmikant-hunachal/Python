```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data={"Salary":[50000,45000,30000,52000,80000,45000,65000,30000,52000,80000,45000,50100],
     "dept":["HR","IT","Non IT","Manager","IT","Non IT","HR","HR","HR","Manager","IT","Non IT"]
     }
df=pd.DataFrame(data)
df
```

```python
Hr_sal=df[df["dept"]=="HR"]["Salary"]
Hr_sal
```

```python
IT_sal=df[df["dept"]=="IT"]["Salary"]
IT_sal
```

```python
ITN_sal=df[df["dept"]=="Non IT"]["Salary"]
ITN_sal
```

```python
Ma_sal=df[df["dept"]=="Manager"]["Salary"]
Ma_sal
```

```python
#first we are creata group of the all salary acc into dept name
plt.boxplot([Hr_sal,IT_sal,ITN_sal,Ma_sal],labels=["HR","IT","Non IT","Manager"])
plt.show()
```

```python
salary_by_dept=df.groupby("dept")["Salary"].sum()
salary_by_dept
```

```python
#group by data represent data in pie chart this data conatin one numeric value and 
#one object type 
plt.pie(salary_by_dept,labels = salary_by_dept.index,autopct="%1.2f",shadow=True,explode=[0.1,0,0.1,0])
plt.axis("equal")
plt.show()
```

```python
hr_mean=sum(Hr_sal)/len(Hr_sal)
IT_mean=sum(IT_sal)/len(IT_sal)
ITN_mean=sum(ITN_sal)/len(ITN_sal)
ma_mean=sum(Ma_sal)/len(Ma_sal)
#first we are find out the mean of the salary acc into dept name
#represention data using bar
plt.bar(["HR","IT","Non IT","Manager"],[hr_mean,IT_mean,ITN_mean,ma_mean],color=["green","yellow","red"])
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data={"Salary":[50000,45000,30000,52000,80000,45000,65000,30000,52000,80000,45000,50100],
     "dept":["HR","IT","Non IT","Manager","IT","Non IT","HR","HR","HR","Manager","IT","Non IT"]
     }
df=pd.DataFrame(data)
df["age"]=[24,25,45,56,47,48,43,42,35,36,38,28]
df["Experience"]=[2.0,2.5,1.5,2.0,1.0,1.5,2.5,3.0,3.5,4.5,5.0,0.5]
df
```

```python
plt.scatter(df["age"],df["Salary"],c=df["dept"].map({"HR":"yellow","IT":"red","Non IT":"green","Manager":"black"}),
            label=["HR","IT","Non IT","Manager"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary VS Age VS Dept")
plt.show()
#above code can not work prpoperly 
```

```python
color={"HR":"yellow","IT":"red","Non IT":"green","Manager":"black"}
for dept,color in color.items():
    df_dept=df[df["dept"]==dept]
    plt.scatter(df_dept["age"],df_dept["Salary"],c=color,label=dept)

plt.grid()
plt.legend()
```

```python
#object orieanted plot. plot more than one graph in one sheet
fig,axs =plt.subplots(1,3,figsize=(15,8))
sort_age=df.sort_values("age")
axs[0].plot(sort_age["age"],df["Salary"],color="red",marker="D",linewidth="2")
axs[0].grid()
axs[0].set_title("Line plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

#histogram
axs[1].hist(df["Salary"],color="skyblue")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Age")
axs[1].set_title("histogram")
#boxplot
axs[2].boxplot(df["Salary"])
axs[2].set_title("Boxplot")
axs[2].set_xlabel("Salary")
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    df["age"],
    df["Salary"],
    df["Experience"],
    s=100,
    c='red'
)

ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_zlabel("Experience")

ax.set_title("3D Scatter Plot")

plt.show()
```

```python
colors = {
    "HR":"yellow",
    "IT":"red",
    "Non IT":"green",
    "Manager":"black"
}

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    df["age"],
    df["Salary"],
    df["Experience"],
    c=df["dept"].map(colors),
    s=df["Experience"]*100
)

ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_zlabel("Experience")

plt.show()
```
